# Log loss function
Let's try to gain an intuition of this function and why it is used.  
When estimating the likelihood of a probabiliy, i.e. as given by the sigmoid function:  

$ \sigma(x) = \frac{1}{1+e^{-x}}$

We need a loss function that will penalise values that are further from the predicted value. The function we'll use is the log loss function, similar to cross-entropy (TBC):  

$ -ylog\hat{y} - (1-y)log(1-\hat{y}) $  

where $y$ is the true probability distribution (ground truth) and $\hat{y}$ is the predicted probability distribution.  
Remembering a probability distribution is a set of probabilities for a given event (TBC).  

So first we define a sequence of numbers in Python, from 0 to 1, representing predicted probabilities. Note, to avoid a division by 0 warning, we start our sequence at 0.001 and finish at 0.999.

```python
import numpy as np
# predicted probabilities
# create a sequence of 100 numbers between 0 and 1
yhat = np.linspace(0.001, 0.999, num=100);
# display
# print(yhat)
```

Next we look at 3 cases, when the true probability is equal to 0, 0.5 and 1.  
Again, to avoid a division by 0 error, we'll use 0.001, 0.5 and 0.999.
We'll plot and comment on each case.  
First, 0.001. We see in the graph that if the ground truth is close to 0, the log loss (or penalty, or cost, or however we define the measure by witch our prediction is penalised) will tend to 0. However, the further our prediction is from 0, the more we are penalised.

```python
import matplotlib.pyplot as plt
# Case 1, ground truth is equal to 0, or very close to
y = 0.001;
logloss = -y*np.log(yhat)-(1-y)*np.log(1-yhat);

plt.plot(yhat, logloss, 'o');
plt.xlabel('Predictions');
plt.ylabel('Log Loss');
plt.title('Log loss for ground truth = 0')
```

Second case, 0.5. We see in the graph that if the ground truth is close to 0.5, the log loss will tend to 0, and values going above or below 0.5 will be equaly penalised.

```python
import matplotlib.pyplot as plt
# Case 2, ground truth is equal to 0.5
y = 0.5;
logloss = -y*np.log(yhat)-(1-y)*np.log(1-yhat);

plt.plot(yhat, logloss, 'o');
plt.xlabel('Predictions');
plt.ylabel('Log Loss');
plt.title('Log loss for ground truth = 0.5')
```

Third case, ground truth equal to 1 (or values very close to 1). We notice a symmetry in relation to the ground truth being equal to 0, where values further from 1 get bigger values, or penalties.

```python
import matplotlib.pyplot as plt
# Case 2, ground truth is equal to 0.5
y = 0.999;
logloss = -y*np.log(yhat)-(1-y)*np.log(1-yhat);

plt.plot(yhat, logloss, 'o');
plt.xlabel('Predictions');
plt.ylabel('Log Loss');
plt.title('Log loss for ground truth = 1')
```

So by looking at the three previous plots, we gain an appreciation of how clever the log loss function is. Imagine if you were given the task to devise an algorithm that would penalise wrong probability predictions such that, the further the prediction was from the ground truth, the higher the penalty would have to be. You could probably do that with some lines of code and get to a reasonable result. Well, that is exactly what the log loss function is doing, in a terse statement.
